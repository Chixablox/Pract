```python
#4

#Так как столбца id нет, то удалять ничего не будем, поэтому переименуем столбец "Temperature (K)"

df = df.rename (
    columns = {
        'Temperature (K)': 'Temperature(K)'
    }
)

df.info()
```

```python
#5

#a
plt.figure(figsize=(10, 6))
plt.hist(df['Absolute magnitude(Mv)'], bins=50)
plt.title('Гистограмма распределения числового признака Absolute magnitude(Mv)')
plt.xlabel('Значение')
plt.ylabel('Частота')
plt.savefig('histogram.png')
plt.show()
```

```python
#b

plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Temperature(K)'])
plt.title('Диаграмма "ящик с усами" для числового признака Temperature(K)')
plt.xlabel('Значение')
plt.savefig('boxplot.png')
plt.show()
```

```python
#c

plt.figure(figsize=(8, 8))
df['Spectral Class'].value_counts().plot.pie(autopct='%0.1f%%')
plt.title('Круговая диаграмма для номинативного признака Spectral Class')
plt.ylabel('')
plt.savefig('pie_chart.png')
plt.show()
```

```python
#d

plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Тепловая карта с корреляцией')
plt.savefig('heatmap.png')
plt.show()
```

```python
#e

plt.figure(figsize=(12, 8))
sns.countplot(x='Star type', hue='Spectral Class', data=df)
plt.title('Диаграмма countplot с группировкой по двум номинативным признакам Star type и Spectral Class')
plt.savefig('countplot.png')
plt.show()
```

```python
#6

df.isna().sum()
```

```python
if df.isnull().any().any():
    for column in df.columns:
        if df[column].dtype == 'int64':
            df[column].fillna(df[column].median(), inplace=True)
        elif df[column].dtype == 'float64':
            df[column].fillna(df[column].mean(), inplace=True)
        else:
            df[column].fillna(df[column].mode().iloc[0], inplace=True)

df.isna().sum()
```

```python
#7

from scipy import stats
sample_size = 200
sample = df['Absolute magnitude(Mv)'].sample(n=sample_size)

statistic, p_value = stats.normaltest(sample)

if p_value < 0.05:
    print("Выборка не является нормально распределенной.")
else:
    print("Выборка может быть нормально распределенной.")
```

```python
#8

columns = ['Star color', 'Spectral Class']

df = pd.get_dummies(df, columns=columns, drop_first= True, dtype = int)

df
```
